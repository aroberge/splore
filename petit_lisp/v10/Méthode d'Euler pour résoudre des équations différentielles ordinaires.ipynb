_Nous présentons la méthode d'Euler, une méthode simple (mais imprécise) pour résoudre numériquement les équations différentielles ordinaires._


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Introduction

Imaginons que nous avons une équation différentielle ordinaire $y' = f(t,y(t))$ avec une condition initiale $y(t_0) = y_0$ et que nous voulions la résoudre numériquement.

## La méthode d'Euler

En premier, nous considérons une approximation pour $f$ basée sur une expansion en série de Taylor autour d'un point dénoté par $t_n$.

$$ y(t_{n+1}) \equiv y(t_n + h) = y(t_{n}) + \frac{y'(t_n)}{1!} h + \frac{y''(t_n)}{2!} h^2 + \frac{y'''(t_n)}{3!} h^3 + \dots $$

où $h := t_{n+1} - t_{n}$ est la distance entre deux points consécutifs. Nous allons uniquement considérer le cas où $h$ est constant à chaque étape (i.e. $\forall n$).

L'approximation la plus simple, connue sous le nom de **méthode explicite d'Euler**, consiste à tronquer la série de Taylor après le terme linéaire, donc ignorant tous les termes d'ordre supérieur.  En substituant cette approximation dans l'équation différentielle $y' = f(t,y(t))$, nous obtenons une équation aux différences du premier ordre
$$
y(t_{n+1}) \approx y(t_n) + hf(t_n, y_n)  + \mathcal{O}(h^2)
$$

et notre séquence d'approximations des points $y_n$ sera obtenue de la façon suivante:

$$
y_{n+1} = y_n + hf(t_n, y_n).
$$

Comme nous négligeons les termes d'ordre supérieur à 1, l'erreur à chaque étape, appelée **erreur locale**, sera de l'ordre de
 $\mathcal{O}(h^2)$. Si $h$ est suffisamment petit, l'erreur cumulative après plusieurs étapes, appelée **erreur globale** sera de l'ordre de $\mathcal{O}(h^1)$.

### Calcul en utilisant le langage Python

Nous allons définir une fonction dans le langage de programmation Python qui démontre la méthode d'Euler.


```python
def euler_eq_diff(fonction, y_0, t, args={}):
    """ Approximation d'Euler pour résoudre des équations différentielles

    Args:
        fonction: La fonction apparaisant dans l'équation différentielle à résoudre. 
        y_0: La condition initale.
        t: une séquence de point pour lesquels on veut trouver y.
        args: arguments supplémentaires si nécessaire

    résultat:
        y: la solution numérique approximative à chaque temps t 
    """

    y = np.zeros([len(t), len(y_0)]) # initialise le vecteur à calculer
    y[0] = y_0

    for i, t_i in enumerate(t[:-1]): # calculer pour chaque intervalle de temps

        h = t[i+1] - t_i  # grandeur de l'intervalle
        y[i+1] = y[i] + h*fonction(t_i, y[i], args)  # approximation d'Euler

    return y
```

## Exemple 1: Une exponentielle

Considérons le cas habituel très simple : 

$$y'(t) = ay(t)$$

avec la valeur initiale $y(0) = 1$ et $a \in \mathbb{R}$. Nous connaissons la solution analytique (exacte) de cette équation, $y = \mathrm{e}^{at}$ ce qui nous permet de vérifier la précision de la méthode. 

Définissons cette équation dans le langage Python.

```python
def exp(t, y, args):
    """ Une fonction exponentielle décrite comme une équation différentielle """
    
    dydt = args['a']*y
    return dydt
```

### Résolution du problème

Fixons notre condition initiale $y_0$ et notre séries de points dans le temps $t_1, t_2 \dots t_N$.

```python
y_0 = np.array([1.]) # condition initiale

N = 10 # nombre de points
t_max = 5.
t = np.linspace(0., t_max, N+1)  # liste des points
```

Quant au facteur $a$, nous allons lui donner la valeur $1$ pour l'instant.

```python
solution_args = {'a':1}
```

Nous sommes mainteant prêts à utiliser la méthode d'Euler pour intégrer cette équation; nous allons également comparer graphiquement avec la solution exacte $y = \mathrm{e}^{t}$.

```python
y = euler_eq_diff(exp, y_0, t, solution_args)  # Calcule la solution avec Euler

plt.plot(t, y[:,0], 'b-o', label='Euler', color="green")
plt.plot(t, np.exp(solution_args['a']*t), 'k--', label='exacte')
plt.xlabel(r'$t$')
plt.ylabel(r'$y(t)$')
plt.legend(loc=2)

plt.show()
```

Bon, avec $N = 10$ points, la solution numérique (en vert) suit la courbe exponentielle mais sous-estime le résultat exact.  Nous savons que les termes "oubliés" de la série de Taylor seront plus petit si on choisi une valeur plus petite d'intervalle $h$.  Vérifions ceci graphiquement.

## Vérification de la précision

Dans le but de vérifier l'effet de choisir un $h$ plus petit, nous allons résoudre le système avec différents nombres de points: $N = 5, 10, 20, 40$ (correspondant à $h = 1, \frac{1}{2}, \tfrac{1}{4}, \tfrac{1}{8}$).

```python
for N in [5, 10, 20, 40]:

    t = np.linspace(0., t_max, N+1) 
    y = euler_eq_diff(exp, y_0, t, solution_args)
    plt.plot(t, y, '-o', label='%d points'%N) 
    
plt.plot(t, np.exp(solution_args['a']*t), 'k--', label='exacte')

plt.xlabel(r'$t$'), plt.ylabel(r'$y$')
plt.legend(loc=2)
print("Graphique comparatif")
```

Ceci est conforme à ce à quoi on s'attendait, ce qui indique que la méthode fonctionne. On peut vérifier que l'erreur globale approximative est de l'ordre $\mathcal{O}(h^1)$ en faisna tun graphique de l'erreur globale au temps $t=5$, donnée par $| \, y_N - e^5 \, |$, en fonction de divers intervalles $h$.

```python
max_N = 16
N = 2**np.arange(2, max_N) # N = 2, 4, 8, ..., 2^max_N

ordre = 1 # pour comparaison visuelle

y_end = np.zeros(len(N))  
intervalle = np.zeros(len(N)) 

for i, N_i in enumerate(N): 

    t = np.linspace(0., t_max, N_i+1)
    y_end[i] = euler_eq_diff(exp, y_0, t, solution_args)[-1]
    
    intervalle[i] = t_max/N_i
    
plt.loglog(intervalle, abs(y_end - np.exp(solution_args['a']*t_max)), 
           'b-o', label='Erreur globale')
plt.loglog(intervalle, intervalle**ordre,'k--', label=r'$h^1$')
plt.xlabel(r'$h$')
plt.legend(loc=2)
print("Graphique logarithmique")
```

Nous voyons que la courbe est linéaire lorsque $h \lesssim 0.1$, ce qui indique que l'ordre est constant. On pourrait essayer de déterminer formellement cette pente mais il suffit de comparer avec la courbe correspondant à la fonction $h^1$ (ligne noire pointillée). Le fait que les deux pentes sont les mêmes indique que l'ordre est $\mathcal{O}(h^1)$.
